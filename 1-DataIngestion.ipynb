## Document Loaders

```python
## 1. Text Loaders
from langchain_community.document_loaders import TextLoader

text = TextLoader('speech.txt')

doc = text.load()

doc
```

```python
## 2. Reading PDF

from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Air conditioner.pdf")

loader.load()
```

```python
## 3. Web Based Loader

from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    [
    'https://www.vskills.in/interview-questions/generative-ai-with-langchain-interview-questions',
    'https://substack.com/home/post/p-151969888?source=queue'
    ],
    
    bs_kwargs=dict(parse_only = bs4.SoupStrainer(
        class_ = ("body markup")
    ))
)

loader.load()
```
